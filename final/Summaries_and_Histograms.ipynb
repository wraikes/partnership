#  Summary Statistics and Histograms

This file shows:
- The summary statistics (count, mean, std, min, max, quantiles) for numeric data.
- The histograms and kernel density estimation when appropriate (std not equal to 0).

To be viewed in HTML.

```python
import os, numpy as np, pandas as pd, seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', 2000)
dmt = pd.read_csv('DMT_merged-DRAFT.csv')
```

```python
dmt.describe()
```

```python
def freq_and_hist(column):
    if dmt[col].dtype == 'float':
        col_na = dmt[col][dmt[col].notnull()]
        if col_na.std() != 0:
            print(dmt[col].describe())
            sns.distplot(col_na)
            plt.show()
```

```python
for col in dmt.columns[0:50]:
    freq_and_hist(col)
    print()
```

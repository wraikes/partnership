#  Frequencies and Summary Statistics

This file shows:
- The summary statistics (count, mean, std, min, max, quantiles) for numeric data.
- The frequency counts for categorical data, with at least one value greater than 1.

To be viewed in HTML.

```python
import os, numpy as np, pandas as pd
dmt = pd.read_csv('DMT_merged-DRAFT.csv')
```

```python
for col in dmt.columns:
    if dmt[col].dtype == 'float':
        print(dmt[col].describe())
    elif dmt[col].dtype == 'object' and dmt[col].value_counts().max() > 1:
        print(dmt[col].value_counts())
    print()
```

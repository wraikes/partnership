### BART

```python
def col_convert_bart(row):
    '''
    Convert a column that contains a string of a list of strings, into a list of floats.  
    
    Inputs
        row: a string of a list of strings.
        
    Output
        returns a list of floats
    
    '''
    if pd.notnull(row):
        return [float(x) for x in row.strip('[]').split(',')]
    
    else:
        return row

def split(row, segment):
    seg = 0, 5
    
    if segment == 2:
        seg = 5, 10
    elif segment == 3:
        seg = 10, 15
    
    if pd.notnull(row):
        return row[seg[0]:seg[1]]
    else:
        return row
```

```python
names = [
    'SAGE_bart_v4_baseline_BART0.25', 
    'SAGE_bart_v4_baseline_BART250.00', 
    'SAGE_bart_v4_21_day_assessment_BART0.25', 
    'SAGE_bart_v4_21_day_assessment_BART250.00'
]

stds = [
    '___pumps_standard_deviation',
    '___pumps_standard_deviation_first_third',
    '___pumps_standard_deviation_second_third',
    '___pumps_standard_deviation_last_third'
]
```

```python
for name in names:
    dmt[name + stds[0]] = dmt[name + '___pumps_per_balloon'].apply(col_convert_bart).apply(np.std)
    dmt[name + stds[1]] = dmt[name + '___pumps_per_balloon'].apply(col_convert_bart).apply(lambda x: x if np.any(pd.isnull(x)) else x[:5]).apply(np.std)
    dmt[name + stds[2]] = dmt[name + '___pumps_per_balloon'].apply(col_convert_bart).apply(lambda x: x if np.any(pd.isnull(x)) else x[5:10]).apply(np.std)
    dmt[name + stds[3]] = dmt[name + '___pumps_per_balloon'].apply(col_convert_bart).apply(lambda x: x if np.any(pd.isnull(x)) else x[10:]).apply(np.std)
```

### Delay Discounting

```python
def col_convert_dd(row, ix=None):
    '''
    Convert a column that contains a string of a list of strings, into a list of floats.
    
    Inputs
        row: a string of a list of strings.
        
    Output
        returns the list of floats
    
    '''
    if pd.notnull(row) and ix:
        return [float(x) for x in row.strip('[]').split(',')][ix]
    
    elif pd.notnull(row):
        return [float(x) for x in row.strip('[]').split(',')]
    
    else:
        return row
```

```python
#Create new column using "last" number in 'choiceArray'.
for col in dmt.columns:
    if 'choiceArray' in col and 'delay' in col:
        col_name = col + '_last'
        dmt[col_name] = dmt[col].apply(lambda x: col_convert_dd(x, -1))
```

```python
#Create new column for times mean.
for col in dmt.columns:
    if 'delay' in col and col.endswith('times'):
        col_name = col + '_mean'
        dmt[col_name] = dmt[col].apply(lambda x: np.mean(col_convert_dd(x)))
```

### Discounting Raw

```python
#Create new column using "last" number in 'choiceArray'/'constantArray'/'variableArray'.
for col in dmt.columns:
    if ('choiceArray' in col or 'constantArray' in col or 'variableArray' in col) and 'SAGE_discounting_raw' in col:
        col_name = col + '_last'
        dmt[col_name] = dmt[col].apply(lambda x: col_convert_dd(x, -1))
```

```python
#Create new column for times mean.
for col in dmt.columns:
    if 'discounting_raw' in col and col.endswith('times'):
        col_name = col + '_mean'
        dmt[col_name] = dmt[col].apply(lambda x: np.mean(col_convert_dd(x)))
```

### Synchronize times

```python
def col_convert(row, option):
    if option == 1:
        if pd.notnull(row):
            return [float(x) if float(x) > 100 else float(x) * 1000 for x in row.strip('[]').split(',')]

        else:
            return row
    else:
        if pd.notnull(row):
            return float(row) if float(row) > 100 else float(row) * 1000
        else:
            return row
```

```python
new_dmt = dmt.copy()

for col in dmt.columns:
    if col.endswith('times'):
        new_dmt[col] = dmt[col].apply(lambda x: col_convert(x, option=1))
    
    elif 'times_v' in col or 'times_mean' in col:
        new_dmt[col] = dmt[col].apply(lambda x: col_convert(x, option=2))
```

```python
labels = [
    '___response_time_mean',
    '___response_time_range',
    '___response_time_std_dev',
    '___response_time_mean_after_one_incorrect',
    '___response_time_mean_after_ten_correct',
    '___response_time_mean_correct',
    '___response_time_mean_incorrect'
]

thirds = [
    '',
    '_first_third',
    '_second_third',
    '_last_third'
]

cols = [
    'SAGE_goNoGo_v2_baseline_go_no_go_stable_stimulus_active_task',
    'SAGE_goNoGo_v2_baseline_go_no_go_variable_stimulus_active_task',
    'SAGE_goNoGo_v2_21_day_assessment_go_no_go_stable_stimulus_active_task',
    'SAGE_goNoGo_v2_21_day_assessment_go_no_go_variable_stimulus_active_task'
]
```

```python
_dmt = new_dmt.copy()

for col in cols:
    for label in labels:
        for third in thirds:
            _dmt[col + label + third] = new_dmt[col + label + third].apply(lambda x: x * 1000 if x > 0 and x < 10 and pd.notnull(x) else x)
```

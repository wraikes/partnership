```python
import pandas as pd, numpy as np
pd.set_option('display.max_columns', 2000)
```

```python
dmt = pd.read_csv('DMT Data Merged DRAFT 10.11.17.csv')
screen = pd.read_csv('screen_output.csv', index_col=0)
call = pd.read_csv('call_output.csv', index_col=0)
bat = pd.read_csv('bat_output.csv', index_col=0)
```

```python
screen.head()
```

```python
new_cols = []

for col in screen.columns:
    new_col = ''
    if col != 'e_id':
        new_col = 'passive_screen_' + col
        new_col = new_col.replace('(', '_')
        new_col = new_col.replace('<', 'less_')
        if new_col.endswith(')'):
            new_col = new_col[:-1]
    else:
        new_col = 'externalId'
    new_cols.append(new_col)

screen.columns = new_cols
screen.head()    
```

```python
bat.head()
```

```python
new_cols = []

for col in bat.columns:
    new_col = ''
    if col != 'e_id':
        new_col = 'passive_bat_' + col
        new_col = new_col.replace('(', '_')
        new_col = new_col.replace('<', 'less_')
        new_col = new_col.replace('>', 'greater_')
        new_col = new_col.replace('%', 'percent')
        if new_col.endswith(')'):
            new_col = new_col[:-1]
    else:
        new_col = 'externalId'
    new_cols.append(new_col)

bat.columns = new_cols
bat.head()    
```

```python
call.head()
```

```python
new_cols = []

for col in call.columns:
    new_col = ''
    if col != 'e_id':
        new_col = 'passive_call_' + col
        new_col = new_col.replace('(', '_')
        if new_col.endswith(')'):
            new_col = new_col[:-1]
    else:
        new_col = 'externalId'
    new_cols.append(new_col)

call.columns = new_cols
call.head()    
```

```python
new_dmt = dmt.merge(bat, on='externalId')
new_dmt = new_dmt.merge(call, on='externalId')
new_dmt = new_dmt.merge(screen, on='externalId')
```

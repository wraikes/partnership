```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt
pd.set_option('display.max_columns', 2000)
%matplotlib inline
dmt = pd.read_csv('../data/DMT_Data_Merged_10.25.17.csv')
```

### Sage

```python
labels = [
    '___number_of_correct_responses', 
    '___number_of_correct_nonresponses',
    '___number_of_incorrect_responses',
    '___number_of_incorrect_nonresponses',
    '___response_time_mean',
    '___response_time_range',
    '___response_time_std_dev',
    '___response_time_mean_after_one_incorrect',
    '___response_time_mean_after_ten_correct',
    '___response_time_mean_correct',
    '___response_time_mean_incorrect'
]

cols = [
    'SAGE_goNoGo_v2_baseline_go_no_go_stable_stimulus_active_task',
    'SAGE_goNoGo_v2_baseline_go_no_go_variable_stimulus_active_task',
    'SAGE_goNoGo_v2_21_day_assessment_go_no_go_stable_stimulus_active_task',
    'SAGE_goNoGo_v2_21_day_assessment_go_no_go_variable_stimulus_active_task'
]
```

```python
for label in labels:
    for col in cols:
        values = dmt[col + label]
        
        if 'time' in label:
            values = values / 1000 #shown in seconds
            plt.xlabel('Seconds')
        
        print(values.describe())
        pd.Series(values).plot.hist(title = label[3:])
        plt.show()
        print()
```

### Inquisit

```python
def time_conv(row):
    if pd.notnull(row) and row != '0:00:00':
        return float(row) / 1000  #Convert to seconds.
    else:
        return np.nan
```

```python
columns = [
    'INQUISIT_gonogo___script.elapsedtime',
    'INQUISIT_gonogo___expressions.errorrate',
    'INQUISIT_gonogo___expressions.errorrate_gotarget',
    'INQUISIT_gonogo___expressions.errorrate_Nogotarget',
    'INQUISIT_gonogo___expressions.meanrt'
]

for column in columns:
    values = dmt[column]
    if 'time' in column or 'meanrt' in column:
        values = pd.Series(values).apply(time_conv)
        plt.xlabel('Seconds')
    
    print(values.describe())
    pd.Series(values).plot.hist(title = column[18:])
    plt.show()
    print()
```

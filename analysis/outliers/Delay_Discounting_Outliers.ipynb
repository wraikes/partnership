```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt
pd.set_option('display.max_columns', 2000)
%matplotlib inline
dmt = pd.read_csv('../../data/final-12-19-17.csv')
```

### Sage

```python
#Outlier analysis for time variables (times shown in seconds).
for col in dmt.columns:
    if 'delay' in col and ('times_v' in col or 'times_mean' in col):
        times = pd.Series([x/1000 for x in dmt[col]], name=dmt[col].name)
        print(times.describe())
        times.plot.hist()
        plt.show()
        print()
```
